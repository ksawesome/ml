### **SECTION 1 Abstract Linear Algebra Basics**

#### **1.1 Vector Spaces Over the Real Numbers**

A **vector space** (or linear space) over the field of real numbers $\mathbb{R}$ is a set $V$ equipped with two operations:

1. **Vector addition**
   $+\colon V \times V \to V$, written $u + v$
2. **Scalar multiplication**
   $\cdot\colon \mathbb{R} \times V \to V$, written $a v$

such that the following axioms hold for all $u, v, w \in V$ and all scalars $a, b \in \mathbb{R}$:

##### **Addition axioms**

1. *(Closure)* $u + v \in V$
2. *(Associativity)* $(u+v)+w = u+(v+w)$
3. *(Commutativity)* $u + v = v + u$
4. *(Zero vector)* There exists $0 \in V$ such that $v + 0 = v$
5. *(Additive inverse)* For each $v$, there exists $-v$ such that $v + (-v) = 0$

##### **Scalar multiplication axioms**

6. *(Compatibility)* $a(bv) = (ab)v$
7. *(Identity)* $1v = v$
8. *(Distributivity over vector addition)* $a(u+v) = au + av$
9. *(Distributivity over scalar addition)* $(a+b)v = av + bv$

A vector space is entirely defined by these axioms; nothing about coordinates, directions, or geometry is assumed.

---

#### **1.2 Subspaces**

A subset $W \subseteq V$ is a **subspace** if:

1. $0 \in W$
2. $u, v \in W \Rightarrow u+v \in W$
3. $u \in W, a \in \mathbb{R} \Rightarrow a u \in W$

Equivalently: closed under addition and scalar multiplication.

##### **Examples**

* Column space of a matrix $A$:
  $\mathrm{Col}(A) = \{Ax : x\in\mathbb{R}^n\}$
* Nullspace of $A$:
  $\mathrm{N}(A) = \{x : Ax=0\}$
* Space of polynomials of degree $\le k$
* Space of continuous functions $C([0,1])$

Subspaces will later become the objects we project onto using matrices like $UU^\top$.

---

#### **1.3 Linear Combinations and Span**

Given vectors $v_1,\dots,v_k \in V$, a **linear combination** is:

$$
a_1 v_1 + \cdots + a_k v_k,\qquad a_i \in \mathbb{R}.
$$

The **span** of these vectors is the set:

$$
\mathrm{span}(v_1,\dots,v_k)
= \{ a_1 v_1 + \cdots + a_k v_k : a_i \in \mathbb{R}\}.
$$

A span is always a subspace of $V$.

---

#### **1.4 Linear Independence**

Vectors $v_1,\dots,v_k$ are **linearly independent** if the only solution to:

$$
a_1 v_1 + \cdots + a_k v_k = 0
$$
is:
$$
a_1 = \cdots = a_k = 0.
$$

If a nontrivial combination equals zero, the set is dependent.

---

#### **1.5 Basis and Dimension**

A set of vectors $v_1,\dots,v_k\in V$ is a **basis** of $V$ if:

1. They are linearly independent
2. They span the entire space

Every vector $v \in V$ can be uniquely expressed as:

$$
v = a_1 v_1 + \cdots + a_k v_k.
$$

The number $k$ is the **dimension** of $V$.

---

#### **1.6 Coordinate Representation**

Once a basis $B = \{v_1,\dots,v_k\}$ is chosen, every vector $v$ corresponds to a unique coordinate vector $[v]_B \in \mathbb{R}^k$:

$$
v = a_1 v_1 + \cdots + a_k v_k \quad \longleftrightarrow \quad
[v]_B = (a_1,\dots,a_k).
$$

Matrices arise exactly from expressing linear maps in coordinates.

---

#### **1.7 Linear Maps (Linear Transformations)**

A function $T : V \rightarrow W$ is **linear** if:

$$
T(u+v) = T(u) + T(v),
\qquad
T(a v) = a T(v)
$$

for all $u,v\in V$ and $a\in\mathbb{R}$.

Once bases are fixed, linear maps correspond to matrices, and matrix multiplication corresponds to composition of linear maps.

This is the bridge from abstract vector spaces → numerical computation.

---

#### **1.8 Special Subspaces in $\mathbb{R}^n$**

When $V=\mathbb{R}^n$, common subspaces include:

* **Column space** of A
* **Row space** (column space of $A^\top$)
* **Nullspace** of A
* **Left nullspace** (nullspace of $A^\top$)

These four subspaces form the **Fundamental Theorem of Linear Algebra** structure and underpin projections, least squares, and SVD.

---

#### **1.9 Key Identities (to remember for later blocks)**

1. Decomposition of any vector $b$ relative to a subspace:
   $$
   b = P b + (I-P)b
   $$

2. Dimension theorem (rank-nullity):
   $$
   \mathrm{rank}(A) + \mathrm{nullity}(A) = n
   $$

3. Column space and nullspace orthogonality:
   $$
   \mathrm{Col}(A) \perp \mathrm{N}(A^\top)
   $$

These will be used directly in Block 2 and in proving projection uniqueness (D2).

### **SECTION 2 Orthogonality, Inner Products & Orthonormal Bases**

---

#### **2.1 Inner Product (Real Vector Spaces)**

An **inner product** on a real vector space $V$ is a function:

$$
\langle \cdot, \cdot \rangle : V \times V \rightarrow \mathbb{R}
$$

satisfying the following properties for all $u,v,w \in V$ and scalar $a\in\mathbb{R}$:

##### **(1) Linearity in the first argument**

$$
\langle u+v, w \rangle = \langle u, w \rangle + \langle v, w \rangle
$$
$$
\langle a u, w \rangle = a, \langle u,w \rangle
$$

##### **(2) Symmetry**

$$
\langle u, v \rangle = \langle v, u \rangle
$$

##### **(3) Positive-definiteness**

$$
\langle v, v \rangle \ge 0 \quad \text{and equals } 0 \text{ iff } v=0
$$

This structure allows us to measure **angles**, **lengths**, and **orthogonality**.

---

#### **2.2 Norm and Distance (Induced by Inner Product)**

Given an inner product, define the **norm**:

$$
|v| = \sqrt{\langle v, v \rangle}
$$

This yields:

* Distance between vectors:
  $$
  d(u,v) = |u - v|
  $$

* Cauchy–Schwarz inequality (must know):
  $$
  |\langle u,v\rangle| \le |u| , |v|
  $$

* Triangle inequality:
  $$
  |u+v| \le |u| + |v|
  $$

These properties define the geometry of the space.

---

#### **2.3 Orthogonality**

Two vectors $u,v \in V$ are **orthogonal** if:
$$
\langle u, v \rangle = 0.
$$

For a set $\{v_1,\dots,v_k\}$:

* Orthogonal ⇔ every pair satisfies $\langle v_i,v_j\rangle = 0$ for $i\neq j$
* If also $|v_i|=1$, the set is **orthonormal**

Orthogonality is essential because it gives **unicity**, **stability**, and **simplicity** in linear decompositions.

---

#### **2.4 Orthonormal Sets and Orthonormal Bases**

A set $\{u_1, \dots, u_k\}$ is **orthonormal** if:
$$
\langle u_i, u_j \rangle = \delta_{ij}.
$$
(where $\delta_{ij}=1$ if $i=j$, else 0)

##### **Orthonormal basis**

An orthonormal set that spans the vector space.

**Key property (must be in notes):**

If $U=[u_1, u_2, \dots, u_k] \in \mathbb{R}^{n\times k}$ is an orthonormal basis of a subspace, then:

$$
U^\top U = I_k.
$$

This property is the backbone of:

* QR decomposition
* Projection ($P = U U^\top$)
* SVD orthogonality
* Numerical stability

---

#### **2.5 Coordinate Representation in Orthonormal Bases**

If $\{u_1, \dots, u_k\}$ is orthonormal, then:

$$
[v]_U = U^\top v
$$

is the coordinate vector of $v$ in this basis.

The reconstruction is:

$$
v = U (U^\top v)
$$

Explanation:

* Projection of $v$ onto span($U$) is simply $UU^\top v$
* No solving linear systems required
* No matrix inverses
* No Gram matrix

This is why orthonormal bases are the “unit testing basis” for ML.

---

#### **2.6 Projection Onto a Subspace (Geometric View)**

Given orthonormal basis $U$ of a subspace $S\subseteq\mathbb{R}^n$:

$$
P = U U^\top
$$

is the orthogonal projector onto $S$.

A vector decomposes uniquely as:
$$
v = Pv + (I-P)v
$$

with:

* $Pv \in S$
* $(I-P)v \in S^\perp$

This division into “signal” and “error” appears in:

* least squares
* PCA
* ridge regression
* spectral clustering
* deep learning initialization theory

---

#### **2.7 Orthogonal Complements**

For a subspace $S$, its **orthogonal complement** is:

$$
S^\perp = \{v : \langle v, s \rangle = 0 ;; \forall s\in S\}
$$

Properties you must include:

* $\mathbb{R}^n = S \oplus S^\perp$
* $\dim S + \dim S^\perp = n$
* $P_{S^\perp} = I - P_S$
* For matrix A:
  $$
  \text{Col}(A) \perp \text{N}(A^\top)
  $$

This feeds into D2 (projection uniqueness) and D8 (least-squares orthogonality).

---

#### **2.8 Orthogonal Matrices**

A matrix $Q\in\mathbb{R}^{n\times n}$ is **orthogonal** if:

$$
Q^\top Q = I
$$

Equivalent statements:

* Columns of Q are orthonormal
* Rows of Q are orthonormal
* Q preserves lengths: $|Qv| = |v|$
* Q preserves inner products:
  $$
  \langle Qv, Qw\rangle = \langle v,w\rangle
  $$

Orthogonal matrices represent **rotations** and **reflections**.
In ML, they maintain numerical stability and structure.

They also underpin:

* QR decomposition
* SVD (U and V always orthogonal)
* PCA basis vectors

---

#### **2.9 Gram–Schmidt (Conceptual Only Here)**

A procedure to turn any linearly independent set $\{a_1,\dots,a_k\}$ into an **orthonormal** one $\{q_1,\dots,q_k\}$.

Conceptual formulas:

$$
\tilde{q}_1 = a_1,\qquad q_1 = \frac{\tilde{q}_1}{|\tilde{q}_1|}
$$

$$
\tilde{q}_i = a_i - \sum_{j < i}(q_j^\top a_i) q_j
$$

$$
q_i = \frac{\tilde{q}_i}{|\tilde{q}_i|}
$$

This feeds directly into:

* D5 (GS correctness proof)
* Coding component of Block 1
* Later QR decomposition and PCA computation

---

#### **2.10 Why Orthonormality Matters (Core Intuition)**

You must include these **five** reasons in your notes:

1. **Projection becomes a single matrix multiply**
   $Pv = UU^\top v$

2. **Coordinates are inner products**
   $[v]_U = U^\top v$

3. **Numerical stability is maximal**
   Orthogonal matrices do not magnify errors.

4. **Diagonalization becomes easy**
   Orthogonal diagonalization in symmetric case (Block 2).

5. **Optimization is easier**
   Gradients in orthonormal bases behave cleanly.

This is foundational: almost every stable algorithm in ML uses orthonormal bases or orthogonal transforms.

### **SECTION 3 Projection Onto a Subspace & Normal Equations**

---

#### **3.1 Orthogonal Projection Onto a Subspace**

Let $U = [u_1 ; u_2 ; \dots ; u_k] \in \mathbb{R}^{n \times k}$ be a matrix whose columns form an **orthonormal set**:

$$
U^\top U = I_k.
$$

Let
$$
S = \mathrm{Col}(U)
$$
be the subspace spanned by these vectors.

Our goal is to derive the **projection matrix** onto $S$, prove its properties, and show it is unique.

---

##### **3.1.1 Derivation of Projection Formula ($P = UU^\top$)**

Any vector $b \in \mathbb{R}^n$ can be decomposed uniquely into:

$$
b = b_\parallel + b_\perp
$$

where:

* $b_\parallel \in S$
* $b_\perp \in S^\perp$
* $\langle b_\parallel, b_\perp\rangle = 0$

Since $S = \text{span}(u_1, \dots, u_k)$, there exist scalars $c_i$ such that:

$$
b_\parallel = \sum_{i=1}^k c_i u_i = Uc,
\quad c = (c_1,\dots,c_k)^\top.
$$

We determine $c$ using orthonormality:

$$
u_j^\top b_\parallel = u_j^\top b \quad \Rightarrow \quad c_j = u_j^\top b.
$$

Thus:

$$
c = U^\top b.
$$

Therefore the projection is:

$$
b_\parallel = U(U^\top b).
$$

Hence the projection operator $P$ is:

$$
\boxed{P = U U^\top}
$$

and:

$$
\boxed{Pb = UU^\top b}
$$

---

##### **3.1.2 Properties of $P = UU^\top$**

You must include all four properties.

###### **(1) P is idempotent**

$$
P^2 = (UU^\top)(UU^\top) = U (U^\top U) U^\top = U I U^\top = UU^\top = P.
$$

###### **(2) P is symmetric**

$$
P^\top = (UU^\top)^\top = U U^\top = P.
$$

###### **(3) Range(P) = S**

$$
Pb = UU^\top b \in \mathrm{Col}(U) = S.
$$

###### **(4) Kernel(P) = $S^\perp$**

If $Pb = 0$, then:

$$
U^\top b = 0 \quad \Rightarrow \quad b\perp u_i ; \forall i.
$$

Thus $b \in S^\perp$.

---

##### **3.1.3 Orthogonality of residual**

For any $b$,

$$
b - P b \in S^\perp.
$$

Proof:

$$
u_j^\top (b - P b) = u_j^\top b - u_j^\top U U^\top b
= u_j^\top b - e_j^\top U^\top b = 0.
$$

Thus residual is orthogonal to every basis vector of $S$.

---

##### **3.1.4 Uniqueness of the Orthogonal Projection**

You must write this exact argument in your notes:

> If $Q$ is any matrix such that:
> $Q^2 = Q$, $Q^\top = Q$, and $\mathrm{Range}(Q) = S$,
> then $Q = UU^\top$.

**Proof sketch for notes:**

1. For any $b \in \mathbb{R}^n$, decompose $b = b_\parallel + b_\perp$ with respect to $S$.
2. Since $Q$ is an orthogonal projector,

   * $Q b_\parallel = b_\parallel$
   * $Q b_\perp = 0$
3. Therefore:
   $$
   Q b = b_\parallel.
   $$
4. But we already showed:
   $$
   UU^\top b = b_\parallel.
   $$
5. Therefore for all $b$,
   $$
   Qb = UU^\top b \quad \Rightarrow \quad Q = UU^\top.
   $$

**Conclusion:**
$$
\boxed{
UU^\top \text{ is the unique orthogonal projector onto } S
}
$$

This derivation is required for D1 and D2.

---

#### **3.2 Least Squares and the Normal Equations**

Given an overdetermined system $Ax=b$ with $A\in\mathbb{R}^{m\times n}$ and $m>n$, exact solutions typically do not exist.
We instead solve:

$$
\min_x |Ax - b|^2.
$$

This finds the **closest vector in the column space of A** to $b$.

---

##### **3.2.1 Geometric Interpretation**

Let:

* $S = \mathrm{Col}(A)$
* $x^*$ = least-squares solution
* $r = b - Ax^*$ = residual

The solution satisfies:

$$
r \in S^\perp
$$

Thus:
$$
A^\top r = 0.
$$

This is the geometric core from which the normal equations follow.

---

##### **3.2.2 Derivation of Normal Equations (Geometric Route)**

From $A^\top r = 0$:

$$
A^\top (b - Ax^*) = 0
$$
$$
A^\top b = A^\top A x^*
$$

Thus:

$$
\boxed{A^\top A x^* = A^\top b}
$$

This characterizes the optimal $x$.

---


The least squares solution is:

$$
x^* = (A^\top A)^{-1} A^\top b
$$
if $A^\top A$ is invertible (full column rank).

Otherwise, the pseudoinverse gives the minimum-norm solution:

$$
x^* = A^+ b.
$$

This connects directly to Module 1 → later Blocks → SVD and pseudoinverse theory.

---

##### **3.2.3 Derivation of Normal Equations (Algebraic Route — Index Notation)**

Objective:

$$
f(x) = |Ax-b|^2 = (Ax-b)^\top(Ax-b)
$$

**Expand component-wise:**

$$
f(x) = \sum_{i=1}^m \left( \sum_{j=1}^n A_{ij} x_j - b_i \right)^2
$$

**Differentiate w.r.t. $x_k$:**

$$
\frac{\partial f}{\partial x_k} = 2 \sum_{i=1}^m A_{ik} \left( \sum_{j=1}^n A_{ij} x_j - b_i \right).
$$

**Set gradient to zero:**

$$
\sum_{i=1}^m A_{ik} \left( \sum_{j=1}^n A_{ij} x_j \right) = \sum_{i=1}^m A_{ik} b_i.
$$

Left-hand side is the $k$-th component of $A^\top A x$.
Right-hand side is $( (A^\top b)_k )$.

**Thus we obtain:**

$$
A^\top A x = A^\top b.
$$

---

##### **3.2.4 Derivation Using Matrix Calculus (Trace Trick)**

**Rewrite objective:**

$$
f(x) = (Ax - b)^\top (Ax - b)
$$

**Expand:**

$$
f(x) = x^\top A^\top A x - 2 b^\top A x + b^\top b.
$$

**Use known derivative:**

$$
\frac{d}{dx} (x^\top M x) = (M + M^\top)x.
$$

Since $A^\top A$ is symmetric:

$$
\frac{d}{dx} (x^\top A^\top A x) = 2 A^\top A x.
$$

**Also:**

$$
\frac{d}{dx} (-2 b^\top A x) = -2 A^\top b.
$$

**Gradient:**

$$
\nabla f(x) = 2A^\top A x - 2A^\top b.
$$

**Set to zero:**

$$
A^\top A x = A^\top b.
$$

Same result.

---

##### **3.2.5 Interpretation of Solution**

The least squares solution is:

$$x^* = (A^\top A)^{-1} A^\top b$$

if $A^\top A$ is invertible (full column rank).

Otherwise, the pseudoinverse gives the minimum-norm solution:

$$x^* = A^+ b.$$

This connects directly to Module 1 → later Blocks → SVD and pseudoinverse theory.

### SECTION 4 Gram–Schmidt & Orthonormal Bases

This section gives: 

1. symbolic algorithms for Classical Gram–Schmidt (CGS) and Modified Gram–Schmidt (MGS). 
2. a careful numerical-caveat discussion explaining why CGS may lose orthogonality and why MGS is more stable. 
3. a formal correctness proof (induction) that Gram–Schmidt produces orthonormal vectors when applied to a linearly independent set.


#### 4.1 Classical Gram–Schmidt (CGS)

**Problem:** Given linearly independent vectors $a_1,\dots,a_k \in \mathbb{R}^n$, construct an orthonormal set $q_1,\dots,q_k$ spanning the same subspace.

**Symbolic formulas (classical GS):**

For $i=1,\dots,k$:
$$
\tilde q_i = a_i - \sum_{j=1}^{i-1} (q_j^\top a_i)\, q_j
$$
$$
q_i = \frac{\tilde q_i}{\|\tilde q_i\|}
$$

**Remarks:**
- The scalar $(q_j^\top a_i)$ is the coordinate (projection coefficient) of $a_i$ along $q_j$.
- After computing $q_i$, it is orthogonal to all previous $q_j,\; j<i$ by construction.


##### 4.1.1 Classical Gram–Schmidt (pseudocode)

```python
# Input: A = [a1, a2, ..., ak] as columns (assumed linearly independent)
# Output: Q with orthonormal columns, and R upper-triangular (optional)

import numpy as np

def classical_gs(A):
    # A: n x k matrix
    n, k = A.shape
    Q = np.zeros((n, k))
    R = np.zeros((k, k))
    
    for i in range(k):
        v = A[:, i].copy()               # start with a_i
        for j in range(i):               # subtract projections onto previous q_j
            r_ji = Q[:, j].T @ A[:, i]
            v = v - r_ji * Q[:, j]
            R[j, i] = r_ji
        
        r_ii = np.linalg.norm(v)
        R[i, i] = r_ii
        Q[:, i] = v / r_ii               # normalize
    
    return Q, R
```


##### 4.1.2 Modified Gram–Schmidt (MGS) — Symbolic Form

**Idea:**  
Instead of subtracting *all* projections from $a_i$ in one vector subtraction (as CGS does),  
MGS removes each projection **immediately**, updating the remaining vectors in-place.

Let $v_i := a_i$ initially.

For $j = 1, \dots, k$:

$$
q_j = \frac{v_j}{\|v_j\|}, \qquad r_{jj} = \|v_j\|
$$

For each $i = j+1, \dots, k$:

$$
r_{j i} = q_j^\top v_i, \qquad
v_i \leftarrow v_i - r_{j i} q_j
$$

This produces the same orthonormal set in exact arithmetic as CGS,  
but with **significantly improved numerical stability**.


##### 4.1.3 Modified Gram–Schmidt (pseudocode)

```python
def modified_gs(A):
    # A: n x k matrix
    n, k = A.shape
    V = A.copy()
    Q = np.zeros((n, k))
    R = np.zeros((k, k))
    
    for j in range(k):
        r_jj = np.linalg.norm(V[:, j])
        Q[:, j] = V[:, j] / r_jj
        R[j, j] = r_jj
        
        for i in range(j+1, k):
            r_ji = Q[:, j].T @ V[:, i]
            R[j, i] = r_ji
            V[:, i] = V[:, i] - r_ji * Q[:, j]   # immediate update
    
    return Q, R
```


##### 4.1.4 How Gram–Schmidt Constructs an Orthonormal Basis

Let the input be linearly independent vectors $a_1,\dots,a_k$.

At step $i$:

1. Remove all components of $a_i$ along the already-constructed orthonormal vectors $q_1,\dots,q_{i-1}$:
   $$
   \tilde q_i = a_i - \sum_{j < i} (q_j^\top a_i) \, q_j.
   $$

2. The vector $\tilde q_i$ is orthogonal to all earlier $q_j$ by construction.

3. Normalize to obtain a unit vector:
   $$
   q_i = \frac{\tilde q_i}{\|\tilde q_i\|}.
   $$

Because the original $a_i$ are linearly independent, no $\tilde q_i$ becomes zero, so the process continues through all $k$ vectors.

The resulting set $\{q_1,\dots,q_k\}$ is orthonormal and spans the same subspace as $\{a_1,\dots,a_k\}$.


#### 4.2 Numerical Caveats — Why Classical GS Loses Orthogonality


##### 4.2.1 Floating-Point Cancellation and Instability in Classical GS

###### 1. Roundoff error
Floating-point arithmetic introduces small relative errors:
$$
fl(x \,\text{op}\, y) = (x \,\text{op}\, y)(1 + \delta), 
\quad |\delta| \lesssim \varepsilon,
$$
where $\varepsilon \approx 10^{-16}$ for double precision.

###### 2. Catastrophic cancellation
When subtracting nearly equal quantities:
$$
v - v_{\parallel},
$$
the leading digits cancel out, leaving a result dominated by roundoff noise.

---

##### 4.2.2 Why Classical GS Suffers

CGS computes:
$$
\tilde q_i = a_i - \sum_{j<i} (q_j^\top a_i)\, q_j.
$$

If $a_i$ is almost in the span of $\{q_1,\dots,q_{i-1}\}$,  
then:
- The sum of projections is nearly equal to $a_i$.
- Their difference $\tilde q_i$ is **a small vector computed by subtracting large ones**.
- This small vector suffers huge relative floating-point error.
- After normalization, its direction becomes inaccurate → orthogonality lost.

Orthogonality error can reach $O(\kappa(A)\varepsilon)$  
where $\kappa(A)$ measures column dependency.

---

##### 4.2.3 Why Modified GS Fixes Most Issues

MGS instead updates *each* remaining vector immediately:

- After forming $q_j$, we do:  
  $v_i \leftarrow v_i - (q_j^\top v_i) q_j$
- These updates involve subtracting vectors of **comparable magnitude**, not a giant accumulated projection.
- Reduces cancellation dramatically.
- Empirically, MGS maintains orthogonality error on the order of  
  $O(\varepsilon)$, near machine precision.

---

##### 4.2.4 Illustrative Example

Let:
$$
u = \begin{bmatrix}1 \\ 0\end{bmatrix},\quad
v = \begin{bmatrix}1 \\ 10^{-12}\end{bmatrix}.
$$

Exact residual after projecting $v$ onto $u$:
$$
v - (u^\top v)u = \begin{bmatrix}0 \\ 10^{-12}\end{bmatrix}.
$$

In floating point:
- $u^\top v \approx 1$ (rounds to 1 exactly)
- Subtraction may yield $[0, 0]^\top$ due to cancellation
- CGS breaks; MGS normalizes residuals sequentially and often avoids total annihilation.

This example demonstrates the **geometric vulnerability** of CGS.


##### 4.2.5 QR Decomposition Perspective

- Classical and Modified GS both compute a QR factorization:
  $$
  A = QR,
  $$
  where $Q$ has orthonormal columns.

- **Householder QR** is more numerically stable than either GS variant.

- GS is primarily used in:
  - simple teaching contexts
  - iterative/streaming QR construction
  - small problems where performance matters more than stability

Monitoring:
$$
\|Q^\top Q - I\|_F
$$
is essential in practice.


#### 4.3 Gram–Schmidt Correctness: Proof of Orthonormality
We prove that the Gram–Schmidt procedure, applied to a linearly independent set, yields orthonormal vectors $q_1,\dots,q_k$.


##### 4.3.1 Proof (by induction)

**Base case:**  
For $i=1$:
$$
q_1 = \frac{a_1}{\|a_1\|}
$$
so $\|q_1\|=1$ trivially.

---

##### 4.3.2 Inductive hypothesis
Assume for some $m \ge 1$ that $q_1,\dots,q_m$ are:
- mutually orthogonal:  
  $$
  q_i^\top q_j = 0, \; i\neq j
  $$
- normalized: $\|q_j\|=1$.

---

##### 4.3.3 Step for $q_{m+1}$

Define:
$$
\tilde{q}_{m+1}
= a_{m+1} - \sum_{j=1}^{m} (q_j^\top a_{m+1}) q_j.
$$

###### 1. Orthogonality  
For any $\ell \le m$:
$$
q_\ell^\top \tilde{q}_{m+1}
= q_\ell^\top a_{m+1}
- \sum_{j=1}^{m} (q_j^\top a_{m+1})(q_\ell^\top q_j)
= q_\ell^\top a_{m+1} - (q_\ell^\top a_{m+1}) = 0.
$$

###### 2. Nonzero residual  
If $\tilde q_{m+1}=0$, then $a_{m+1}$ lies in the span of $q_1,\dots,q_m$,  
contradicting independence of $a_1,\dots,a_k$.

###### 3. Normalization  
Define:
$$
q_{m+1} = \frac{\tilde q_{m+1}}{\|\tilde q_{m+1}\|}.
$$
Then $\|q_{m+1}\|=1$ and $q_{m+1}^\top q_j = 0$ for all $j \le m$.

---

##### 4.3.4 Conclusion
By induction, Gram–Schmidt yields orthonormal vectors:
$$
q_i^\top q_j = 0\; (i\neq j), \qquad \|q_i\| = 1.
$$

Thus the algorithm is algebraically correct.


#### 4.4 Final Remarks

- Algebraic correctness holds in exact arithmetic; floating-point behavior requires Modified GS or Householder QR for stability.
- The proofs above justify the structure of the QR factorization and the projection formulas that appear later in PCA, SVD, and least squares.

### SECTION 5 The 8 Required Derivations

In this section, we work through the eight core derivations that establish the geometric and algebraic foundations of projections, least squares, orthogonality, and Gram–Schmidt.
Each derivation is written so it can be referenced later when studying PCA, spectral methods, pseudoinverses, and optimization.

 #### **D1 — Projection Formula Correctness**

Let $U \in \mathbb{R}^{n \times k}$ be a matrix whose columns are **orthonormal**, i.e.

$$
U^\top U = I_k.
$$

Define the matrix

$$
P = U U^\top.
$$

We want to prove that $P$ is **the orthogonal projector** onto the subspace $\operatorname{Col}(U)$.

This requires showing:

1. $P^2 = P$ (idempotence)
2. $P^\top = P$ (symmetry)
3. $Pb \in \operatorname{Col}(U)$ for all $b \in \mathbb{R}^n$
4. $b - Pb \perp \operatorname{Col}(U)$
5. Therefore, $P$ is the unique orthogonal projector onto $\operatorname{Col}(U)$



##### **1. Idempotence: $P^2 = P$**

Compute:

$$
P^2 = (U U^\top)(U U^\top)
= U (U^\top U) U^\top
= U I_k U^\top
= U U^\top
= P.
$$

Thus $P$ is idempotent.



##### **2. Symmetry: $P^\top = P$**

Take the transpose:

$$
P^\top = (U U^\top)^\top = (U^\top)^\top U^\top = U U^\top = P.
$$

Hence $P$ is symmetric.



##### **3. $Pb \in \operatorname{Col}(U)$**

Let $b \in \mathbb{R}^n$. Then

$$
Pb = U U^\top b.
$$

Since $U^\top b \in \mathbb{R}^k$ is a coefficient vector, the expression $U(U^\top b)$ is simply a **linear combination of the columns of $U$**.

Therefore,

$$
Pb \in \operatorname{Col}(U).
$$



##### **4. Orthogonality of the residual: $(b - Pb) \perp \operatorname{Col}(U)$**

Take any vector $u$ in $\operatorname{Col}(U)$.
Since the columns of $U$ span the subspace, any $u$ can be written as $u = U c$ for some $c \in \mathbb{R}^k$.

Compute inner product:

$$
u^\top (b - Pb)
= (Uc)^\top (b - UU^\top b)
= c^\top U^\top b - c^\top U^\top U U^\top b
$$

Use $U^\top U = I$:

$$
= c^\top U^\top b - c^\top I U^\top b
= c^\top U^\top b - c^\top U^\top b
= 0.
$$

Thus,

$$
u^\top (b - Pb) = 0 \quad \forall u \in \operatorname{Col}(U)
$$

so

$$
b - Pb \perp \operatorname{Col}(U).
$$



##### **5. Conclusion: $P$ is the orthogonal projector onto $\operatorname{Col}(U)$**

We have shown:

* $P$ is idempotent ($P^2 = P$)
* $P$ is symmetric ($P^\top = P$)
* $Pb$ lies within $\operatorname{Col}(U)$
* $b - Pb$ is orthogonal to the subspace

These are exactly the defining properties of an orthogonal projection operator.
By standard theorems in linear algebra, an operator is the **unique** orthogonal projector onto a subspace if and only if it is **symmetric and idempotent**.

Therefore:

$$
\boxed{
P = U U^\top \text{ is the unique orthogonal projection onto } \operatorname{Col}(U).
}
$$

#### **D2 — Uniqueness of the Orthogonal Projection**

**Claim.**
If a matrix $P\in\mathbb{R}^{n\times n}$ satisfies
$$P^2 = P \quad\text{and}\quad P^\top = P,$$
then $P$ is the orthogonal projector onto the subspace $\operatorname{Range}(P)$. Moreover, this orthogonal projector is unique.

We prove the following steps:

1. $\operatorname{Range}(P)$ is the subspace onto which $P$ projects (i.e. $P$ acts as identity on $\operatorname{Range}(P)$).
2. $\operatorname{Null}(P)=\operatorname{Range}(P)^\perp$.
3. No other symmetric idempotent matrix can project orthogonally onto the same subspace; hence the projector is unique.



##### 1. $P$ acts as the identity on $\operatorname{Range}(P)$

Let $s\in\operatorname{Range}(P)$. By definition there exists $x$ with $s = P x$. Then
$$P s = P(Px) = P^2 x = P x = s,$$
so $P$ fixes every vector in $\operatorname{Range}(P)$. Hence $P$ maps $\operatorname{Range}(P)$ to itself and acts as the identity there. Therefore $\operatorname{Range}(P)$ is the candidate subspace onto which $P$ projects.



##### 2. $\operatorname{Null}(P)=\operatorname{Range}(P)^\perp$

We show two inclusions.

###### (i) $\operatorname{Null}(P)\subseteq\operatorname{Range}(P)^\perp$

Let $v\in\operatorname{Null}(P)$, so $Pv=0$. Take any $s\in\operatorname{Range}(P)$. There exists $w$ with $s = Pw$. Then
$$s^\top v = (Pw)^\top v = w^\top P^\top v = w^\top P v = w^\top 0 = 0,$$
where we used $P^\top = P$. Thus $s$ is orthogonal to $v$. As this holds for every $s\in\operatorname{Range}(P)$, we have $v\in\operatorname{Range}(P)^\perp$.

###### (ii) $\operatorname{Range}(P)^\perp \subseteq \operatorname{Null}(P)$

Let $v\in\operatorname{Range}(P)^\perp$. Consider the vector $Pv$. Since $Pv\in\operatorname{Range}(P)$, and $v$ is orthogonal to every vector in $\operatorname{Range}(P)$, we have
$$(Pv)^\top v = 0.$$

Compute the squared norm of $Pv$:
$$|Pv|^2 = (Pv)^\top (Pv) = v^\top P^\top P v.$$

Using $P^\top = P$ and $P^2 = P$,
$$|Pv|^2 = v^\top P P v = v^\top P v = (Pv)^\top v = 0.$$

Thus $|Pv|^2 = 0$ and therefore $Pv=0$. Hence $v\in\operatorname{Null}(P)$.

Combining (i) and (ii) gives
$$\operatorname{Null}(P) = \operatorname{Range}(P)^\perp.$$



##### 3. Uniqueness of the orthogonal projector onto $\operatorname{Range}(P)$

Suppose $Q$ is another matrix with $Q^2=Q$, $Q^\top=Q$, and $\operatorname{Range}(Q)=\operatorname{Range}(P)=:S$. We show $Q=P$.

Every $x\in\mathbb{R}^n$ admits the orthogonal decomposition with respect to $S$:
$$x = s + t,\qquad s\in S, \quad t\in S^\perp.$$

(Explicitly, take $s = Px$ and $t = x - Px$; we have shown $Px\in S$ and $x-Px\in S^\perp$.)

Because $P$ and $Q$ are projectors onto $S$, they must both act as the identity on $S$ and send $S^\perp$ to $0$. Therefore for our decomposition,
$$P x = P(s+t) = Ps + Pt = s + 0 = s,$$

and similarly
$$Q x = Q(s+t) = Qs + Qt = s + 0 = s.$$

Thus $Px = Qx$ for every $x\in\mathbb{R}^n$, which implies $P = Q$.

Hence the symmetric idempotent matrix projecting onto $S$ is unique.



##### Conclusion

A matrix $P$ with $P^2=P$ and $P^\top=P$:

* Projects every vector $x$ onto the subspace $S=\operatorname{Range}(P)$ via $Px$
* Has nullspace equal to $S^\perp$
* Is the unique orthogonal projector onto $S$

Therefore symmetric idempotence characterizes orthogonal projection matrices and guarantees their uniqueness.

#### **D3 — Normal Equations (Index Notation)**

**Claim.**
For an overdetermined system $Ax = b$ with $A \in \mathbb{R}^{m \times n}$ (where $m > n$), the least-squares solution is characterized by:
$$\frac{\partial}{\partial x} \|Ax - b\|^2 = 0 \quad \Rightarrow \quad A^\top A x = A^\top b.$$

We derive this by computing the derivative component-wise using index notation.



##### Setup

Define the objective function:
$$f(x) = \|Ax - b\|^2.$$

Expanding the squared norm:
$$f(x) = (Ax - b)^\top (Ax - b) = \sum_{i=1}^m (Ax - b)_i^2 = \sum_{i=1}^m \left(\sum_{j=1}^n A_{ij} x_j - b_i\right)^2.$$



##### Step 1: Differentiate w.r.t. $x_k$

Fix an index $k \in \{1, \ldots, n\}$ and compute:
$$\frac{\partial f}{\partial x_k} = \frac{\partial}{\partial x_k} \sum_{i=1}^m \left(\sum_{j=1}^n A_{ij} x_j - b_i\right)^2.$$

Using the chain rule:
$$\frac{\partial f}{\partial x_k} = \sum_{i=1}^m 2\left(\sum_{j=1}^n A_{ij} x_j - b_i\right) \cdot \frac{\partial}{\partial x_k}\left(\sum_{j=1}^n A_{ij} x_j - b_i\right).$$

Since $\frac{\partial}{\partial x_k}\left(\sum_{j=1}^n A_{ij} x_j\right) = A_{ik}$:
$$\frac{\partial f}{\partial x_k} = 2 \sum_{i=1}^m A_{ik} \left(\sum_{j=1}^n A_{ij} x_j - b_i\right).$$



##### Step 2: Set gradient to zero

The optimality condition is $\nabla f(x) = 0$, i.e., $\frac{\partial f}{\partial x_k} = 0$ for all $k$:
$$\sum_{i=1}^m A_{ik} \left(\sum_{j=1}^n A_{ij} x_j - b_i\right) = 0 \quad \text{for all } k = 1, \ldots, n.$$



##### Step 3: Recognize the matrix form

The $k$-th component equation is:
$$\sum_{i=1}^m A_{ik} \sum_{j=1}^n A_{ij} x_j = \sum_{i=1}^m A_{ik} b_i.$$

The left-hand side is the $k$-th component of $A^\top A x$:
$$(A^\top A x)_k = \sum_{i=1}^m A_{ik} \sum_{j=1}^n A_{ij} x_j.$$

The right-hand side is the $k$-th component of $A^\top b$:
$$(A^\top b)_k = \sum_{i=1}^m A_{ik} b_i.$$



##### Step 4: Write in matrix form

Since this holds for every $k = 1, \ldots, n$, we have:
$$\boxed{A^\top A x = A^\top b}.$$

These are the **normal equations** for the least-squares problem.



##### Solution

If $A$ has full column rank (i.e., $\operatorname{rank}(A) = n$), then $A^\top A$ is invertible, and the unique least-squares solution is:
$$\boxed{x^* = (A^\top A)^{-1} A^\top b}.$$

Otherwise, the set of solutions forms an affine subspace, and the minimum-norm solution is given by the pseudoinverse:
$$\boxed{x^* = A^+ b},$$
where $A^+$ is the Moore–Penrose pseudoinverse.



##### Remarks

1. **Geometric interpretation:** The residual $r = b - Ax^*$ is orthogonal to the column space of $A$: $A^\top r = 0$.

2. **Numerical stability:** Computing $x^* = (A^\top A)^{-1} A^\top b$ directly can suffer from ill-conditioning if $\operatorname{cond}(A^\top A) = [\operatorname{cond}(A)]^2$ is large. QR decomposition or SVD are preferred in practice.

3. **Connection to projections:** The solution $x^*$ projects $b$ onto $\operatorname{Col}(A)$ via $Ax^* = A(A^\top A)^{-1} A^\top b$. The projector is $P = A(A^\top A)^{-1} A^\top$.

#### **D4 — Normal Equations (Matrix Calculus Derivation)**

**Claim.**
For an overdetermined system $Ax = b$ with $A \in \mathbb{R}^{m \times n}$ (where $m > n$), the least-squares solution is characterized by using the matrix norm expansion and the derivative identity for quadratic forms.

We derive the normal equations using:
$$\|Ax - b\|^2 = (Ax - b)^\top (Ax - b)$$
and the identity:
$$\frac{d}{dx}(x^\top B x) = (B + B^\top) x.$$



##### Step 1: Expand the squared norm

$$\|Ax - b\|^2 = (Ax - b)^\top (Ax - b).$$

Expanding the product:
$$\|Ax - b\|^2 = x^\top A^\top A x - 2 b^\top A x + b^\top b.$$

Define the objective function:
$$f(x) = x^\top A^\top A x - 2 b^\top A x + b^\top b.$$



##### Step 2: Apply the quadratic form derivative identity

We use the given identity for the derivative of a quadratic form. Since $A^\top A$ is symmetric,
$$(A^\top A)^\top = A^\top A.$$

By the identity:
$$\frac{d}{dx}(x^\top A^\top A x) = (A^\top A + (A^\top A)^\top) x = (A^\top A + A^\top A) x = 2 A^\top A x.$$

For the linear term:
$$\frac{d}{dx}(-2 b^\top A x) = -2 A^\top b.$$

For the constant term:
$$\frac{d}{dx}(b^\top b) = 0.$$



##### Step 3: Compute the gradient

$$\nabla f(x) = 2 A^\top A x - 2 A^\top b.$$



##### Step 4: Set gradient to zero

The optimality condition is $\nabla f(x) = 0$:
$$2 A^\top A x - 2 A^\top b = 0.$$

Dividing by 2:
$$\boxed{A^\top A x = A^\top b}.$$

These are the **normal equations** for the least-squares problem.



##### Solution

If $A$ has full column rank (i.e., $\operatorname{rank}(A) = n$), then $A^\top A$ is invertible, and the unique least-squares solution is:
$$\boxed{x^* = (A^\top A)^{-1} A^\top b}.$$

Otherwise, the set of solutions forms an affine subspace, and the minimum-norm solution is given by the pseudoinverse:
$$\boxed{x^* = A^+ b},$$
where $A^+$ is the Moore–Penrose pseudoinverse.



##### Remarks

1. **Equivalence with D3:** This derivation gives the same result as D3 (index notation) but uses matrix calculus directly, avoiding component-wise differentiation.

2. **Geometric interpretation:** The residual $r = b - Ax^*$ satisfies $A^\top r = 0$, meaning it is orthogonal to the column space of $A$.

3. **Hessian verification:** The Hessian of $f(x)$ is $H = 2 A^\top A$. For full column rank $A$, this is positive definite, confirming $x^*$ is a minimum.

4. **Practical computation:** Computing $x^* = (A^\top A)^{-1} A^\top b$ directly can suffer from ill-conditioning when $\operatorname{cond}(A^\top A) = [\operatorname{cond}(A)]^2$ is large. QR decomposition or SVD are preferred in practice.

#### **D5 — Gram–Schmidt Orthogonality Proof**

**Claim.**
When the Gram–Schmidt algorithm is applied to a linearly independent set $\{a_1, \ldots, a_k\}$, the resulting vectors $\{q_1, \ldots, q_k\}$ satisfy:
$$q_i^\top q_j = 0 \quad \text{for all } i \neq j.$$

We prove this by induction using the projection definition.



##### **Base Case ($i = 1$)**

For $i = 1$, there are no earlier vectors, so the orthogonality condition is vacuous. We have:
$$q_1 = \frac{a_1}{\|a_1\|}.$$

The base case holds trivially.



##### **Inductive Hypothesis**

Assume that for some $m \ge 1$, the vectors $q_1, \ldots, q_m$ constructed by Gram–Schmidt are pairwise orthogonal:
$$q_i^\top q_j = 0 \quad \text{for all } 1 \le i < j \le m.$$



##### **Inductive Step: Prove Orthogonality of $q_{m+1}$**

By the Gram–Schmidt algorithm, the residual is:
$$\tilde{q}_{m+1} = a_{m+1} - \sum_{j=1}^{m} (q_j^\top a_{m+1})\, q_j.$$

This can be rewritten using the projection onto the span of $\{q_1, \ldots, q_m\}$:
$$\tilde{q}_{m+1} = a_{m+1} - P_m a_{m+1},$$
where $P_m = \sum_{j=1}^{m} q_j q_j^\top$ (orthogonal projector onto span$(q_1, \ldots, q_m)$).

By construction, $\tilde{q}_{m+1}$ is orthogonal to the column space of the matrix $Q_m = [q_1 \mid \cdots \mid q_m]$:
$$Q_m^\top \tilde{q}_{m+1} = Q_m^\top a_{m+1} - Q_m^\top P_m a_{m+1}.$$

Since $P_m = Q_m Q_m^\top$:
$$Q_m^\top \tilde{q}_{m+1} = Q_m^\top a_{m+1} - Q_m^\top Q_m Q_m^\top a_{m+1}.$$

By the inductive hypothesis, $Q_m^\top Q_m = I_m$ (the columns are already orthonormal):
$$Q_m^\top \tilde{q}_{m+1} = Q_m^\top a_{m+1} - I_m Q_m^\top a_{m+1} = 0.$$

Therefore, $\tilde{q}_{m+1}$ is orthogonal to all $q_j$ for $j = 1, \ldots, m$.

After normalization:
$$q_{m+1} = \frac{\tilde{q}_{m+1}}{\|\tilde{q}_{m+1}\|},$$

the vector $q_{m+1}$ inherits this orthogonality:
$$q_{m+1}^\top q_j = 0 \quad \text{for all } j = 1, \ldots, m.$$



##### **Conclusion**

By induction, the Gram–Schmidt algorithm produces an orthonormal set:
$$\boxed{q_i^\top q_j = 0 \quad \text{for all } i \neq j, \quad \text{and} \quad \|q_i\| = 1}.$$

This proof is fundamental to understanding why Gram–Schmidt works and why it produces an orthogonal basis from any linearly independent set.

#### **D6 — Gram–Schmidt Numerical Stability Analysis**

**Goal:**
Explain how subtracting two nearly equal vectors in floating-point arithmetic leads to catastrophic cancellation and precision loss. This derivation note demonstrates the fundamental mechanism behind why Classical Gram–Schmidt fails numerically.



##### **Setup: Floating-Point Representation**

In double precision (IEEE 754), machine epsilon is:
$$\varepsilon \approx 2.22 \times 10^{-16}.$$

Any real number $x$ is represented as:
$$\text{fl}(x) = x(1 + \delta), \quad |\delta| \le \varepsilon_{\text{rel}},$$
where $\varepsilon_{\text{rel}}$ is the relative error.



##### **Example: Projection and Subtraction**

Consider two vectors in $\mathbb{R}^2$:
$$u = \begin{bmatrix}1 \\ 0\end{bmatrix}, \quad v = \begin{bmatrix}1 \\ 10^{-12}\end{bmatrix}.$$

We want to compute the residual after projecting $v$ onto $u$:
$$r = v - (u^\top v) u.$$



##### **Exact Computation**

**Step 1: Inner product**

$$u^\top v = 1 \cdot 1 + 0 \cdot 10^{-12} = 1.$$

**Step 2: Scaled projection**

$$(u^\top v) u = 1 \cdot \begin{bmatrix}1 \\ 0\end{bmatrix} = \begin{bmatrix}1 \\ 0\end{bmatrix}.$$

**Step 3: Residual (exact)**

$$r = \begin{bmatrix}1 \\ 10^{-12}\end{bmatrix} - \begin{bmatrix}1 \\ 0\end{bmatrix} = \begin{bmatrix}0 \\ 10^{-12}\end{bmatrix}.$$

The second component has magnitude $10^{-12}$, which is well above machine epsilon.



##### **Floating-Point Computation: Where It Fails**

Now consider what happens with finite precision. The issue arises not in the inner product $u^\top v$ (which equals 1 exactly), but in the subtraction.

**The magnitude of the vectors:**
- $|v| \approx 1 + O(10^{-24})$ (nearly 1)
- $|(u^\top v)u| = 1$ (exactly)

**Subtraction of nearby quantities:**

When we compute:
$$r = v - (u^\top v)u,$$

we are subtracting two vectors of comparable magnitude ($\sim 1$), and the result is a small vector ($\sim 10^{-12}$).

**Relative error analysis:**

The relative error in the subtraction is:
$$\frac{|\text{fl}(r) - r|}{|r|} \approx \frac{\varepsilon \cdot \max(|v|, |(u^\top v)u|)}{|r|} = \frac{\varepsilon \cdot 1}{10^{-12}} = 10^{12} \varepsilon \approx 10^{-4}.$$

This is catastrophic: we lose about **4 significant digits** in the result!



##### **Mechanism of Cancellation**

In standard floating-point arithmetic:

1. **Input precision:** Both $v$ and $(u^\top v)u$ are represented with relative error $\sim \varepsilon$.

2. **Magnitude of operands:** Both have size $\sim 1$, so absolute error is $\sim \varepsilon$.

3. **Magnitude of result:** $|r| \sim 10^{-12}$, much smaller than the inputs.

4. **Relative error amplification:**
   $$\text{rel.error}(r) = \frac{\text{abs.error in result}}{\text{magnitude of result}} \approx \frac{\varepsilon}{10^{-12}} = 10^{12} \varepsilon.$$

The **cancellation** occurs because:
- The leading digits of $v$ and $(u^\top v)u$ are nearly identical.
- When subtracted, these leading digits cancel out.
- What remains is a result dominated by rounding errors in the least significant digits.



##### **Implication for Gram–Schmidt**

In Classical Gram–Schmidt, at step $i$:
$$\tilde q_i = a_i - \sum_{j<i} (q_j^\top a_i) q_j.$$

If $a_i$ is nearly in the span of $\{q_1, \ldots, q_{i-1}\}$ (which happens when columns are nearly dependent):
- $\sum_{j<i} (q_j^\top a_i) q_j$ has magnitude comparable to $|a_i|$.
- Their difference $\tilde q_i$ is small.
- The small result $\tilde q_i$ inherits the amplified relative error from the subtraction.
- After normalization, this error becomes directional error in $q_i$.
- The computed $q_i$ is no longer orthogonal to earlier vectors.



##### **Why Modified Gram–Schmidt Avoids This**

Modified Gram–Schmidt updates vectors immediately:

For $i = j+1, \ldots, k$:
$$v_i \leftarrow v_i - (q_j^\top v_i) q_j.$$

**Key difference:**
- Each update removes a **single projection**, not an accumulated sum.
- $|q_j^\top v_i| \ll |v_i|$ (for nearly dependent columns), so the update is small.
- The updated $v_i$ remains large relative to the subtraction error.
- Relative error stays bounded: $\sim O(\varepsilon)$.



##### **Numerical Verification**

For the example above with Classical GS:

| Quantity | Exact Value | Computed (fp64) | Relative Error |
|----------|-------------|-----------------|----------------|
| $u^\top v$ | $1.0$ | $1.0$ | $0$ |
| $(u^\top v)u$ | $(1, 0)$ | $(1.0, 0.0)$ | $\sim \varepsilon$ |
| $v - (u^\top v)u$ | $(0, 10^{-12})$ | $(0, 10^{-12} \pm 10^{-4})$ | $\sim 10^{-4}$ |

The computed second component has absolute error $\sim 10^{-4}$, making the result unreliable for orthogonality checks.



##### **Conclusion**

$$\boxed{\text{Subtracting nearly equal vectors amplifies relative errors by factor} \sim \frac{\text{size of operands}}{\text{size of result}}.}$$

For Gram–Schmidt with nearly dependent columns, this amplification factor can reach $O(\kappa(A))$, where $\kappa(A)$ is the condition number.

**Solution:** Modified Gram–Schmidt sidesteps this by making smaller, sequential updates that maintain comparable magnitudes throughout, keeping errors at the level of machine epsilon rather than amplifying them.

#### **D7 — Projection Error Decomposition Identity**

**Claim.**
For any vector $b \in \mathbb{R}^n$ and an orthonormal matrix $U \in \mathbb{R}^{n \times k}$ with $U^\top U = I_k$, the vector $b$ can be decomposed into two orthogonal components whose squared norms satisfy:

$$\boxed{\|b\|^2 = \|U^\top b\|^2 + \|(I - UU^\top)b\|^2}.$$

This is the Pythagorean theorem for orthogonal decomposition. We prove orthogonality, then algebraic correctness, and finally interpret the energy decomposition.



##### **Part 1: Orthogonality of the Components**

Define:
$$b_\parallel = UU^\top b \quad \text{(projection onto} \, \operatorname{Col}(U))$$
$$b_\perp = (I - UU^\top)b \quad \text{(orthogonal residual)}.$$

We first show that these two components are orthogonal:
$$b_\parallel^\top b_\perp = 0.$$

**Proof:**

Compute the inner product:
$$b_\parallel^\top b_\perp = (UU^\top b)^\top (I - UU^\top)b = b^\top UU^\top (I - UU^\top)b.$$

Expand:
$$b^\top UU^\top (I - UU^\top)b = b^\top U (U^\top - U^\top UU^\top) b.$$

Use $U^\top U = I_k$:
$$b^\top U (U^\top - U^\top I U^\top) b = b^\top U (U^\top - U^\top U^\top) b = b^\top U(U^\top - U^\top) b = 0.$$

Therefore:
$$\boxed{b_\parallel \perp b_\perp}.$$



##### **Part 2: Algebraic Verification (Pythagorean Theorem)**

Since $b_\parallel$ and $b_\perp$ are orthogonal, and their sum is $b$:
$$b = b_\parallel + b_\perp,$$

the Pythagorean theorem gives:
$$\|b\|^2 = \|b_\parallel\|^2 + \|b_\perp\|^2.$$

We now verify this algebraically by expanding each term.

**Expand $\|b_\parallel\|^2$:**

$$\|b_\parallel\|^2 = \|UU^\top b\|^2 = (UU^\top b)^\top (UU^\top b) = b^\top UU^\top UU^\top b.$$

Using $U^\top U = I_k$:
$$\|b_\parallel\|^2 = b^\top U (U^\top U) U^\top b = b^\top U I U^\top b = b^\top UU^\top b.$$

Alternatively, we can express this using coordinates. Let $c = U^\top b \in \mathbb{R}^k$. Then:
$$b_\parallel = Uc, \quad \|b_\parallel\|^2 = \|Uc\|^2 = (Uc)^\top(Uc) = c^\top U^\top U c = c^\top I c = \|c\|^2 = \|U^\top b\|^2.$$

Therefore:
$$\boxed{\|b_\parallel\|^2 = \|U^\top b\|^2}.$$

**Expand $\|b_\perp\|^2$:**

$$\|b_\perp\|^2 = \|(I - UU^\top)b\|^2 = ((I - UU^\top)b)^\top ((I - UU^\top)b).$$

Expand:
$$\|b_\perp\|^2 = b^\top (I - UU^\top)^\top (I - UU^\top) b.$$

Since $I$ and $UU^\top$ are both symmetric:
$$\|b_\perp\|^2 = b^\top (I - UU^\top)(I - UU^\top) b = b^\top (I - 2UU^\top + UU^\top UU^\top) b.$$

Use $UU^\top UU^\top = (UU^\top)^2 = UU^\top$ (idempotence from D1):
$$\|b_\perp\|^2 = b^\top (I - 2UU^\top + UU^\top) b = b^\top (I - UU^\top) b.$$

Therefore:
$$\boxed{\|b_\perp\|^2 = \|(I - UU^\top)b\|^2}.$$

**Combine to verify the decomposition:**

$$\|b_\parallel\|^2 + \|b_\perp\|^2 = b^\top UU^\top b + b^\top(I - UU^\top)b$$
$$= b^\top (UU^\top + I - UU^\top)b = b^\top I b = b^\top b = \|b\|^2.$$

Thus:
$$\boxed{\|b\|^2 = \|U^\top b\|^2 + \|(I - UU^\top)b\|^2}.$$



##### **Part 3: Energy Interpretation**

The decomposition has a powerful interpretation in terms of **energy** or **variance** captured by the subspace.

**Signal and Noise:**
- $b_\parallel = UU^\top b$ is the component of $b$ **in** the subspace $\operatorname{Col}(U)$ — the "signal".
- $b_\perp = (I - UU^\top)b$ is the component **orthogonal** to the subspace — the "noise" or "error".

**Energy conservation:**
$$\text{Total energy} = \text{Energy in subspace} + \text{Energy orthogonal to subspace}$$
$$\|b\|^2 = \|U^\top b\|^2 + \|(I - UU^\top)b\|^2.$$

**Energy captured by the subspace:**

The fraction of energy (variance) captured by the $k$-dimensional subspace $\operatorname{Col}(U)$ is:
$$\text{Fraction captured} = \frac{\|U^\top b\|^2}{\|b\|^2}.$$

This quantity measures how "aligned" $b$ is with the subspace. Values close to 1 mean $b$ lies mostly in $\operatorname{Col}(U)$.

**Applications:**

1. **Principal Component Analysis (PCA):**
   When $U$ contains the top $k$ eigenvectors of the covariance matrix, this ratio measures how much of the data variance is explained by the first $k$ principal components.

2. **Least Squares:**
   The ratio $\frac{\|Uc\|^2}{\|b\|^2}$ quantifies the goodness-of-fit: how much of $b$ is explained by the model $\operatorname{Col}(A)$.

3. **Compression and Approximation:**
   Keeping only $b_\parallel$ discards energy $\|b_\perp\|^2$. The approximation error is $\|(I - UU^\top)b\|$.



##### **Connection to Projections**

Recall from D1 that $P = UU^\top$ is the orthogonal projector onto $\operatorname{Col}(U)$.

The decomposition is:
$$b = Pb + (I-P)b,$$

where:
- $Pb$ lies in $\operatorname{Col}(U)$
- $(I-P)b$ lies in $\operatorname{Col}(U)^\perp$

The identity states that these two orthogonal components satisfy the Pythagorean theorem:
$$\|b\|^2 = \|Pb\|^2 + \|(I-P)b\|^2.$$



##### **Conclusion**

The projection error decomposition identity is fundamental to understanding:
- How orthogonal projections partition vectors
- How much energy is captured versus lost
- The geometry underlying PCA, least squares, and spectral methods

$$\boxed{\|b\|^2 = \|U^\top b\|^2 + \|(I - UU^\top)b\|^2 \quad \text{(Pythagorean theorem for orthogonal decomposition)}}$$

#### **D8 — Orthogonality of Residual in Least Squares**

**Claim.**
If $x^*$ minimizes the least-squares objective $\|Ax - b\|^2$, then the residual vector:
$$r^* = b - Ax^*$$
is orthogonal to the column space of $A$. Equivalently:
$$\boxed{A^\top (b - Ax^*) = 0} \quad \Rightarrow \quad A^\top r^* = 0.$$

We prove this by setting the gradient to zero, then interpret the result geometrically.



##### **Part 1: Derivative Argument (First-Order Optimality)**

**Setup:**

Define the objective function:
$$f(x) = \|Ax - b\|^2 = (Ax - b)^\top (Ax - b).$$

Expand:
$$f(x) = x^\top A^\top A x - 2 b^\top A x + b^\top b.$$

**Compute the gradient:**

Taking the derivative with respect to $x$:
$$\nabla f(x) = 2 A^\top A x - 2 A^\top b.$$

**Set gradient to zero (first-order optimality condition):**

At the minimum, $\nabla f(x^*) = 0$:
$$2 A^\top A x^* - 2 A^\top b = 0.$$

Dividing by 2:
$$A^\top A x^* = A^\top b.$$

**Rearrange to reveal the residual:**

Subtract $A^\top b$ from both sides:
$$A^\top A x^* - A^\top b = 0.$$

Factor:
$$A^\top (Ax^* - b) = 0.$$

Multiply by $-1$:
$$\boxed{A^\top (b - Ax^*) = 0}.$$

This is the **normal equation**, and it tells us that the residual $r^* = b - Ax^*$ is orthogonal to the row space of $A$, which is equivalent to being orthogonal to the column space of $A^\top$.



##### **Part 2: Geometric Interpretation**

The condition $A^\top r^* = 0$ has a profound geometric meaning.

**Understanding $A^\top r^* = 0$:**

Recall that for any vector $v$:
$$A^\top v = 0 \quad \Leftrightarrow \quad v \in \operatorname{N}(A^\top) = \operatorname{Col}(A)^\perp.$$

(The nullspace of $A^\top$ is the orthogonal complement of the column space of $A$.)

Therefore:
$$A^\top (b - Ax^*) = 0 \quad \Leftrightarrow \quad (b - Ax^*) \in \operatorname{Col}(A)^\perp.$$

This means:
$$\boxed{r^* = b - Ax^* \text{ is orthogonal to } \operatorname{Col}(A)}.$$

**Decomposition:**

The geometry of least squares can be visualized as follows:

- **$\operatorname{Col}(A)$:** The subspace spanned by the columns of $A$ (the "feasible region").
- **$b$:** An arbitrary target vector (typically not in $\operatorname{Col}(A)$).
- **$Ax^*$:** The closest point in $\operatorname{Col}(A)$ to $b$ (the orthogonal projection of $b$ onto $\operatorname{Col}(A)$).
- **$r^* = b - Ax^*$:** The residual, pointing perpendicular to $\operatorname{Col}(A)$.

The decomposition is:
$$b = \underbrace{Ax^*}_{\in \operatorname{Col}(A)} + \underbrace{r^*}_{\in \operatorname{Col}(A)^\perp}.$$

By D7 (projection error decomposition):
$$\|b\|^2 = \|Ax^*\|^2 + \|r^*\|^2.$$



##### **Part 3: Verification Using Projection**

We can verify this result using the projection framework from D1.

**Projection onto $\operatorname{Col}(A)$:**

Assume $A$ has full column rank. The orthogonal projector onto $\operatorname{Col}(A)$ is:
$$P = A(A^\top A)^{-1}A^\top.$$

The projection of $b$ onto $\operatorname{Col}(A)$ is:
$$P b = A(A^\top A)^{-1}A^\top b.$$

This is exactly $Ax^*$ where $x^* = (A^\top A)^{-1}A^\top b$.

**Residual is orthogonal to $\operatorname{Col}(A)$:**

By D1 property 4, for any projection $P$:
$$(I - P)b \in \operatorname{Col}(A)^\perp.$$

The residual is:
$$r^* = b - Ax^* = b - Pb = (I - P)b.$$

Therefore, $r^*$ lies in the orthogonal complement of $\operatorname{Col}(A)$, which means:
$$A^\top r^* = 0.$$



##### **Part 4: Error Minimization Intuition**

Why does making the residual orthogonal to $\operatorname{Col}(A)$ minimize the error?

**Answer:** By the Pythagorean theorem (D7), we can decompose:
$$\|b - Ax\|^2 = \|b - Ax^*\|^2 + \|(Ax^* - Ax)\|^2 + 2(b-Ax^*)^\top(Ax^* - Ax).$$

For any $x \neq x^*$, we have $Ax - Ax^* \neq 0$. If $(b - Ax^*) \perp \operatorname{Col}(A)$, then:
$$(b - Ax^*)^\top(Ax - Ax^*) = 0$$
because $Ax - Ax^* \in \operatorname{Col}(A)$.

Thus:
$$\|b - Ax\|^2 = \|b - Ax^*\|^2 + \|A(x - x^*)\|^2 > \|b - Ax^*\|^2$$

for any $x \neq x^*$. This proves $x^*$ minimizes the least-squares error.



##### **Part 5: Connection to Normal Equations**

The orthogonality condition $A^\top r^* = 0$ is equivalent to the normal equations:

Starting from $A^\top(b - Ax^*) = 0$:
$$A^\top b - A^\top A x^* = 0$$
$$\boxed{A^\top A x^* = A^\top b}.$$

This is the fundamental normal equation, which characterizes the least-squares solution.



##### **Conclusion**

The orthogonality of the least-squares residual is the **geometric essence** of least squares:

$$\boxed{\text{The optimal } x^* \text{ makes the residual } r^* \text{ orthogonal to the column space of } A.}$$

This principle connects:
- **Optimization:** Setting $\nabla f = 0$ gives $A^\top r^* = 0$.
- **Geometry:** The residual $r^*$ lies in $\operatorname{Col}(A)^\perp$.
- **Linear algebra:** The normal equations $A^\top A x = A^\top b$.
- **Signal processing:** The error is uncorrelated with any component in the model space.

This condition is used throughout applied mathematics: in regression, signal processing, machine learning, and the foundations of many numerical algorithms.

### Computation 1 — Classical Gram–Schmidt (CGS)

**Goal:** implement the classical Gram–Schmidt algorithm for a full-rank matrix `A ∈ ℝ^{n×k}` and measure:
- orthogonality error: $\|Q^\top Q - I\|_F$,
- reconstruction error: $\|A - Q R\|_F$,
- runtime (single-run wall time measured with `time.perf_counter()`).

**Requirements / constraints**
1. Use `float64` everywhere. Set `np.random.seed(0)` at start of the cell to ensure reproducibility for this computation.
2. Implement function `classical_gs(A)` exactly (no `np.linalg.qr` calls). It must:
   - return `(Q, R)` with shapes `(n,k)` and `(k,k)`,
   - raise `np.linalg.LinAlgError` if a diagonal `R[j,j] == 0` (rank deficiency).
3. Provide helper functions:
   - `orthogonality_error(Q)` → Frobenius norm of `Q.T @ Q - I`
   - `reconstruction_error(A, Q, R)` → Frobenius norm of `A - Q @ R`
   - `time_function(func, *args, repeats=5)` → mean wall time and std.
4. Run a baseline experiment:
   - `n = 50`, `k = 20`, `A = np.random.randn(n, k)`
   - Run `classical_gs(A)` once and report orthogonality & reconstruction errors and time for a single run.
   - Repeat the run `repeats = 10` times to collect mean ± std of errors and runtime.
5. Perform an assertion check: reconstruction error must be ≤ `1e-9` (for float64, random well-conditioned A). If assertion fails, print diagnostics (max abs of A - Q @ R and `Q.T @ Q`).
6. Keep outputs succinct: print a small summary table with metrics (mean ± std).


````python
# Computation 1 — Classical Gram–Schmidt implementation & baseline experiment
import numpy as np
import time

np.random.seed(0)  # reproducibility for this computation

def classical_gs(A):
    """
    Classical Gram-Schmidt.
    Args:
        A: (n, k) ndarray, n >= k, assumed full column rank.
    Returns:
        Q: (n, k) ndarray with orthonormal columns (float64)
        R: (k, k) upper-triangular ndarray (float64)
    Raises:
        np.linalg.LinAlgError if a diagonal R[j,j] == 0 (rank deficiency).
    Complexity: O(n k^2)
    """
    A = np.asarray(A, dtype=np.float64)
    n, k = A.shape
    Q = np.zeros((n, k), dtype=np.float64)
    R = np.zeros((k, k), dtype=np.float64)
    for j in range(k):
        v = A[:, j].copy()
        for i in range(j):
            R[i, j] = Q[:, i].dot(v)
            v = v - R[i, j] * Q[:, i]
        R[j, j] = np.linalg.norm(v)
        if R[j, j] == 0.0:
            raise np.linalg.LinAlgError(f"Rank deficiency encountered at column {j} in classical_gs")
        Q[:, j] = v / R[j, j]
    return Q, R

def orthogonality_error(Q):
    M = Q.T @ Q - np.eye(Q.shape[1], dtype=np.float64)
    return np.linalg.norm(M, ord='fro')

def reconstruction_error(A, Q, R):
    return np.linalg.norm(A - Q @ R, ord='fro')

def time_function(func, *args, repeats=5):
    times = []
    for _ in range(repeats):
        t0 = time.perf_counter()
        func(*args)
        t1 = time.perf_counter()
        times.append(t1 - t0)
    times = np.array(times, dtype=np.float64)
    return float(times.mean()), float(times.std())

# Baseline parameters
n, k = 50, 20
A = np.random.randn(n, k)

# Single run (show outputs)
Q, R = classical_gs(A)
ort_err = orthogonality_error(Q)
rec_err = reconstruction_error(A, Q, R)
print("Single run results:")
print(f"  Orthogonality error ||Q^T Q - I||_F = {ort_err:.3e}")
print(f"  Reconstruction error ||A - Q R||_F = {rec_err:.3e}")

# Repeated runs for statistics
repeats = 10
ort_errs = []
rec_errs = []
times = []
for _ in range(repeats):
    t0 = time.perf_counter()
    Qr, Rr = classical_gs(A)
    t1 = time.perf_counter()
    ort_errs.append(orthogonality_error(Qr))
    rec_errs.append(reconstruction_error(A, Qr, Rr))
    times.append(t1 - t0)

ort_errs = np.array(ort_errs)
rec_errs = np.array(rec_errs)
times = np.array(times)

print("\nSummary over {} runs:".format(repeats))
print(f"  Orthogonality error mean ± std: {ort_errs.mean():.3e} ± {ort_errs.std():.3e}")
print(f"  Reconstruction error mean ± std: {rec_errs.mean():.3e} ± {rec_errs.std():.3e}")
print(f"  Runtime (s) mean ± std: {times.mean():.3e} ± {times.std():.3e}")

# Assertion on reconstruction quality
tol = 1e-9
if rec_errs.mean() > tol:
    print("\n[ASSERTION WARNING] Reconstruction error mean exceeds tolerance.")
    max_abs_diff = np.max(np.abs(A - Qr @ Rr))
    print(f"  max |A - QR| = {max_abs_diff:.3e}")
    print("  Q^T Q - I (smallest/ largest entries):", np.min(Qr.T @ Qr - np.eye(k)), np.max(Qr.T @ Qr - np.eye(k)))
else:
    print(f"\nAssertion passed: mean reconstruction error ≤ {tol:.1e}")

# Export CGS results for comparison in Computation 2
ort_errs_cgs = ort_errs.copy()
rec_errs_cgs = rec_errs.copy()
times_cgs = times.copy()
````

````python
# Computation 2 — Modified Gram–Schmidt implementation & baseline experiment
import numpy as np
import time

np.random.seed(0)  # reproducibility for this computation

def modified_gs(A):
    """
    Modified Gram-Schmidt.
    Args:
        A: (n, k) ndarray, n >= k, assumed full column rank.
    Returns:
        Q: (n, k) ndarray with orthonormal columns (float64)
        R: (k, k) upper-triangular ndarray (float64)
    Raises:
        np.linalg.LinAlgError if a diagonal R[j,j] == 0 (rank deficiency).
    Complexity: O(n k^2)
    """
    A = np.asarray(A, dtype=np.float64)
    n, k = A.shape
    V = A.copy()
    Q = np.zeros((n, k), dtype=np.float64)
    R = np.zeros((k, k), dtype=np.float64)
    for j in range(k):
        R[j, j] = np.linalg.norm(V[:, j])
        if R[j, j] == 0.0:
            raise np.linalg.LinAlgError(f"Rank deficiency encountered at column {j} in modified_gs")
        Q[:, j] = V[:, j] / R[j, j]
        for i in range(j+1, k):
            R[j, i] = Q[:, j].dot(V[:, i])
            V[:, i] = V[:, i] - R[j, i] * Q[:, j]
    return Q, R

# Reuse helper functions (assumes they exist in the notebook; redefine if needed)
def orthogonality_error(Q):
    M = Q.T @ Q - np.eye(Q.shape[1], dtype=np.float64)
    return np.linalg.norm(M, ord='fro')

def reconstruction_error(A, Q, R):
    return np.linalg.norm(A - Q @ R, ord='fro')

def time_function(func, *args, repeats=5):
    times = []
    for _ in range(repeats):
        t0 = time.perf_counter()
        func(*args)
        t1 = time.perf_counter()
        times.append(t1 - t0)
    times = np.array(times, dtype=np.float64)
    return float(times.mean()), float(times.std())

# Baseline parameters (should match Computation 1)
n, k = 50, 20
A = np.random.randn(n, k)

# Single run (show outputs)
Q_mgs, R_mgs = modified_gs(A)
ort_err_mgs = orthogonality_error(Q_mgs)
rec_err_mgs = reconstruction_error(A, Q_mgs, R_mgs)
print("Single run (MGS) results:")
print(f"  Orthogonality error ||Q^T Q - I||_F = {ort_err_mgs:.3e}")
print(f"  Reconstruction error ||A - Q R||_F = {rec_err_mgs:.3e}")

# Repeated runs for statistics
repeats = 10
ort_errs = []
rec_errs = []
times = []
for _ in range(repeats):
    t0 = time.perf_counter()
    Qr, Rr = modified_gs(A)
    t1 = time.perf_counter()
    ort_errs.append(orthogonality_error(Qr))
    rec_errs.append(reconstruction_error(A, Qr, Rr))
    times.append(t1 - t0)

ort_errs = np.array(ort_errs)
rec_errs = np.array(rec_errs)
times = np.array(times)

print("\nMGS Summary over {} runs:".format(repeats))
print(f"  Orthogonality error mean ± std: {ort_errs.mean():.3e} ± {ort_errs.std():.3e}")
print(f"  Reconstruction error mean ± std: {rec_errs.mean():.3e} ± {rec_errs.std():.3e}")
print(f"  Runtime (s) mean ± std: {times.mean():.3e} ± {times.std():.3e}")

# Comparison to CGS if CGS results exist in namespace
try:
    ort_errs_cgs  # reference to ensure variable exists
    rec_errs_cgs
    times_cgs
    print("\nComparison (MGS vs CGS):")
    print(f"  Orthogonality error (MGS mean) = {ort_errs.mean():.3e}, (CGS mean) = {ort_errs_cgs.mean():.3e}")
    print(f"  Reconstruction error (MGS mean) = {rec_errs.mean():.3e}, (CGS mean) = {rec_errs_cgs.mean():.3e}")
    print(f"  Runtime (MGS mean) = {times.mean():.3e}, (CGS mean) = {times_cgs.mean():.3e}")
except NameError:
    print("\nCGS baseline results not found in namespace; skipping direct numeric comparison.")

# Assertion: MGS orthogonality <= CGS orthogonality for baseline A (if CGS data present)
try:
    assert ort_errs.mean() <= ort_errs_cgs.mean() + 1e-14
    print("\nAssertion passed: MGS orthogonality ≤ CGS orthogonality (within tolerance).")
except NameError:
    pass
except AssertionError:
    print("\n[ASSERTION WARNING] MGS orthogonality is larger than CGS on baseline A.")
    print(f"  MGS ort mean = {ort_errs.mean():.3e}; CGS ort mean = {ort_errs_cgs.mean():.3e}")
    # Diagnostic prints
    print("  Sample Q^T Q - I (MGS first):")
    print((Qr.T @ Qr - np.eye(k))[:3, :3])
    print("  Sample Q^T Q - I (CGS first):")
    print((Q.T @ Q - np.eye(k))[:3, :3])
````

````python
# Computation 3 — Orthogonality diagnostics
import numpy as np

def orthogonality_diagnostics(Q):
    """
    Compute orthogonality diagnostics for Q.
    Returns a dict with:
      - ort_err: Frobenius norm ||Q^T Q - I||_F
      - spec_err: spectral norm (largest singular value) of (Q^T Q - I)
      - max_abs: maximum absolute entry of (Q^T Q - I)
      - mean_offdiag: mean absolute value of off-diagonal entries
      - max_offdiag_idx: tuple (i,j) of largest abs off-diagonal entry
      - cond_Q: condition number of Q (np.linalg.cond) as heuristic
    """
    Q = np.asarray(Q, dtype=np.float64)
    k = Q.shape[1]
    M = Q.T @ Q - np.eye(k, dtype=np.float64)
    ort_err = np.linalg.norm(M, ord='fro')
    # spectral norm via svd of M (or np.linalg.norm(M,2))
    try:
        spec_err = np.linalg.norm(M, ord=2)
    except Exception:
        spec_err = float(np.linalg.svd(M, compute_uv=False)[0]) if M.size else 0.0
    max_abs = float(np.max(np.abs(M)))
    # off-diagonals mask
    offdiag_mask = ~np.eye(k, dtype=bool)
    offdiag_vals = np.abs(M[offdiag_mask]) if k > 1 else np.array([0.0])
    mean_offdiag = float(offdiag_vals.mean()) if offdiag_vals.size else 0.0
    if offdiag_vals.size:
        max_off_val = float(offdiag_vals.max())
        # find index of max absolute off-diagonal
        idx_flat = int(np.argmax(np.abs(M) * offdiag_mask))
        i_idx = idx_flat // k
        j_idx = idx_flat % k
        max_off_idx = (i_idx, j_idx)
    else:
        max_off_val = 0.0
        max_off_idx = (0, 0)
    # condition number heuristic
    try:
        cond_Q = float(np.linalg.cond(Q))
    except Exception:
        cond_Q = float('inf')
    return {
        'ort_err': ort_err,
        'spec_err': spec_err,
        'max_abs': max_abs,
        'mean_offdiag': mean_offdiag,
        'max_offdiag_val': max_off_val,
        'max_offdiag_idx': max_off_idx,
        'cond_Q': cond_Q
    }

# Acquire Q candidates. Prefer variables from previous computations if present.
candidates = {}
# Try names commonly used in prior cells
names_to_try = [('Q','cgs'), ('Q_mgs','mgs'), ('Q_mgs','mgs_alt'), ('Qr','cgs_last'), ('Qr','mgs_last')]
for varname, tag in names_to_try:
    if varname in globals():
        try:
            candidates[tag] = globals()[varname]
        except Exception:
            pass

# If none found, compute Q via classical_gs on baseline A (assumes A exists)
if not candidates:
    try:
        Qtemp, _ = classical_gs(A)
        candidates['cgs_generated'] = Qtemp
        # also generate MGS for comparison
        try:
            Qtemp2, _ = modified_gs(A)
            candidates['mgs_generated'] = Qtemp2
        except Exception:
            pass
    except Exception:
        raise RuntimeError("No Q matrices found in namespace and failed to generate from A. Ensure previous cells ran.")

# Run diagnostics for each candidate
ort_diagnostics = {}
for tag, Qmat in candidates.items():
    diag = orthogonality_diagnostics(Qmat)
    ort_diagnostics[tag] = diag
    print(f"[{tag}] ort_err={diag['ort_err']:.3e}, spec_err={diag['spec_err']:.3e}, max_abs={diag['max_abs']:.3e}, mean_offdiag={diag['mean_offdiag']:.3e}, cond_Q={diag['cond_Q']:.3e}")

# Warnings and detailed diagnostics
for tag, diag in ort_diagnostics.items():
    if diag['ort_err'] > 1e-8:
        print(f"WARNING: [{tag}] Frobenius orthogonality error {diag['ort_err']:.3e} > 1e-8")
    if diag['mean_offdiag'] > 1e-6:
        i, j = diag['max_offdiag_idx']
        print(f"NOTE: [{tag}] largest off-diagonal at index (i={i}, j={j}) = {diag['max_offdiag_val']:.3e}")

# Store diagnostics in globals for later cells
globals()['ort_diagnostics'] = ort_diagnostics
````

### Computation 4 — Reconstruction Error

**Goal:** compute and analyze reconstruction error for a factorization `A ≈ Q R` produced by CGS/MGS (or any QR-like factorization).

We will compute robust diagnostics for the reconstruction and provide actionable checks:

1. Implement `reconstruction_diagnostics(A, Q, R)` that returns:
   - `rec_err_fro`: Frobenius norm `||A - Q R||_F`
   - `rec_err_rel`: relative Frobenius error `||A - Q R||_F / ||A||_F`
   - `max_abs_diff`: maximum absolute entry `max_{i,j} |A_{ij} - (Q R)_{ij}|`
   - `mean_abs_diff`: mean absolute entry error
   - `residual_norm_cols`: column-wise residual norms `||A[:,j] - (Q R)[:,j]||_2` as a vector
   - `orthogonality_check`: reuse `orthogonality_error(Q)` to correlate with reconstruction
2. Reuse `A` and candidate `Q,R` pairs from prior computations:
   - CGS: `(Q, R)` (commonly `Q, R` or `Qr, Rr`)
   - MGS: `(Q_mgs, R_mgs)`
   - If not available, compute both via `classical_gs(A)` and `modified_gs(A)`.
3. For each candidate, compute diagnostics and print a concise table.
4. Assertions:
   - `rec_err_rel` should be ≤ `1e-12` for well-conditioned synthetic A (floating point limits permitting). If not, print diagnostics.
   - `max_abs_diff` should be small (e.g., ≤ `1e-9`); otherwise print the largest entry index and value.
5. Store diagnostics in `reconstruction_diagnostics_all = {'cgs':..., 'mgs':...}`.
6. Keep outputs succinct and numeric (scientific notation). Do not plot in this cell.


````python
# Computation 4 — Reconstruction diagnostics
import numpy as np

def reconstruction_diagnostics(A, Q, R):
    """
    Diagnostics for reconstructing A from Q and R.
    Returns:
      - rec_err_fro: ||A - Q R||_F
      - rec_err_rel: ||A - Q R||_F / ||A||_F
      - max_abs_diff: max abs entry of residual
      - max_abs_idx: index (i,j) of max abs entry
      - mean_abs_diff: mean abs entry of residual
      - residual_norm_cols: vector of 2-norms for each column residual
      - orthogonality_err: orthogonality_error(Q)
    """
    A = np.asarray(A, dtype=np.float64)
    Q = np.asarray(Q, dtype=np.float64)
    R = np.asarray(R, dtype=np.float64)
    Resid = A - (Q @ R)
    rec_err_fro = float(np.linalg.norm(Resid, ord='fro'))
    normA = float(np.linalg.norm(A, ord='fro'))
    rec_err_rel = rec_err_fro / normA if normA != 0 else float('inf')
    abs_res = np.abs(Resid)
    max_abs_diff = float(abs_res.max())
    max_idx_flat = int(np.argmax(abs_res))
    i_idx = max_idx_flat // A.shape[1]
    j_idx = max_idx_flat % A.shape[1]
    mean_abs_diff = float(abs_res.mean())
    residual_norm_cols = np.linalg.norm(Resid, axis=0)
    orthogonality_err = orthogonality_error(Q)
    return {
        'rec_err_fro': rec_err_fro,
        'rec_err_rel': rec_err_rel,
        'max_abs_diff': max_abs_diff,
        'max_abs_idx': (i_idx, j_idx),
        'mean_abs_diff': mean_abs_diff,
        'residual_norm_cols': residual_norm_cols,
        'orthogonality_err': orthogonality_err
    }

# Collect Q,R candidates (reuse variables if present)
candidates = {}
if 'Q' in globals() and 'R' in globals():
    candidates['cgs'] = (globals()['Q'], globals()['R'])
if 'Q_mgs' in globals() and 'R_mgs' in globals():
    candidates['mgs'] = (globals()['Q_mgs'], globals()['R_mgs'])
# fallback: generate if missing
if not candidates:
    try:
        Qc, Rc = classical_gs(A)
        candidates['cgs_generated'] = (Qc, Rc)
    except Exception as e:
        print("Failed to generate CGS Q,R:", e)
    try:
        Qm, Rm = modified_gs(A)
        candidates['mgs_generated'] = (Qm, Rm)
    except Exception as e:
        print("Failed to generate MGS Q,R:", e)

reconstruction_diagnostics_all = {}
for tag, (Qcand, Rcand) in candidates.items():
    diag = reconstruction_diagnostics(A, Qcand, Rcand)
    reconstruction_diagnostics_all[tag] = diag
    print(f"[{tag}] rec_err_fro={diag['rec_err_fro']:.3e}, rec_err_rel={diag['rec_err_rel']:.3e}, max_abs={diag['max_abs_diff']:.3e}, mean_abs={diag['mean_abs_diff']:.3e}, ort_err={diag['orthogonality_err']:.3e}")

# Assertions & detailed diagnostics
tol_rel = 1e-12
tol_abs = 1e-9
for tag, diag in reconstruction_diagnostics_all.items():
    if diag['rec_err_rel'] > tol_rel:
        print(f"WARNING [{tag}]: relative reconstruction error {diag['rec_err_rel']:.3e} > {tol_rel:.1e}")
    if diag['max_abs_diff'] > tol_abs:
        i, j = diag['max_abs_idx']
        print(f"WARNING [{tag}]: max abs residual {diag['max_abs_diff']:.3e} at index (i={i}, j={j})")

# Store for downstream use
globals()['reconstruction_diagnostics_all'] = reconstruction_diagnostics_all
````

````python
# Computation 6 — Runtime comparison across problem sizes
import numpy as np
import time
from math import log
np.random.seed(2)

def time_function_once(func, *args):
    t0 = time.perf_counter()
    func(*args)
    t1 = time.perf_counter()
    return t1 - t0

def time_method_on_A(method_name, A, repeats=5):
    times = []
    for _ in range(repeats):
        if method_name == 'cgs':
            t = time_function_once(classical_gs, A)
        elif method_name == 'mgs':
            t = time_function_once(modified_gs, A)
        elif method_name == 'householder':
            t = time_function_once(np.linalg.qr, A, 'reduced')  # returns Q,R
        else:
            raise ValueError("Unknown method")
        times.append(t)
    times = np.array(times, dtype=np.float64)
    return float(times.mean()), float(times.std()), times

# Experiment parameters
k_list = [20, 50, 100, 200]
repeats = 5
runtime_results = []

for k in k_list:
    n = 5 * k
    A = np.random.randn(n, k)  # same A used for all methods for fairness
    condA = np.linalg.cond(A)
    entry = {'k': k, 'n': n, 'cond': float(condA), 'methods': {}}
    # time CGS
    mean_cgs, std_cgs, times_cgs = time_method_on_A('cgs', A, repeats=repeats)
    entry['methods']['cgs'] = {'mean': mean_cgs, 'std': std_cgs, 'times': times_cgs.tolist()}
    # time MGS
    mean_mgs, std_mgs, times_mgs = time_method_on_A('mgs', A, repeats=repeats)
    entry['methods']['mgs'] = {'mean': mean_mgs, 'std': std_mgs, 'times': times_mgs.tolist()}
    # time Householder (np.linalg.qr)
    mean_hh, std_hh, times_hh = time_method_on_A('householder', A, repeats=repeats)
    entry['methods']['householder'] = {'mean': mean_hh, 'std': std_hh, 'times': times_hh.tolist()}
    runtime_results.append(entry)
    print(f"k={k:4d}, n={n:4d}, cond(A)={condA:.3e} | CGS {mean_cgs:.3e}±{std_cgs:.1e}s | MGS {mean_mgs:.3e}±{std_mgs:.1e}s | HH {mean_hh:.3e}±{std_hh:.1e}s")

# store runtime_results globally
globals()['runtime_results'] = runtime_results

# Empirical scaling: fit slope log(time) ~ slope * log(k) + intercept for each method
def fit_scaling(runtime_results, method):
    ks = np.array([e['k'] for e in runtime_results], dtype=np.float64)
    times_mean = np.array([e['methods'][method]['mean'] for e in runtime_results], dtype=np.float64)
    # Only keep positive times
    mask = times_mean > 0
    ks = ks[mask]
    times_mean = times_mean[mask]
    if len(ks) < 2:
        return None
    logk = np.log(ks)
    logt = np.log(times_mean)
    slope, intercept = np.polyfit(logk, logt, 1)
    return slope, intercept

for method in ['cgs', 'mgs', 'householder']:
    fit = fit_scaling(runtime_results, method)
    if fit is not None:
        slope, intercept = fit
        print(f"Empirical scaling for {method}: time ∝ k^{slope:.2f} (slope={slope:.2f})")
    else:
        print(f"Insufficient data to fit scaling for {method}")
````

**Key Observations**

1. **Householder QR is fastest by a large margin**: At $k=200$, Householder (`np.linalg.qr`) runs in **16 ms** compared to CGS at **78 ms** and MGS at **118 ms** — approximately **5–7× faster** than the Gram–Schmidt implementations.

2. **CGS is faster than MGS**: CGS consistently outperforms MGS by ~20–50% across all problem sizes, because MGS performs additional memory writes to update remaining columns at each step.

3. **Empirical scaling is subquadratic in k**: All three methods show scaling around $k^{2.2}$–$k^{2.4}$, which is lower than the theoretical $O(nk^2) = O(k^3)$ expected when $n = 5k$. This suggests:
   - Cache effects and vectorized operations reduce effective complexity.
   - LAPACK's Householder QR benefits from highly optimized BLAS routines.

4. **Condition numbers are excellent**: All test matrices have $\kappa(A) \approx 2.3$–$2.6$, confirming well-conditioned inputs where all methods are numerically stable.

5. **Variance is low for MGS and Householder**: Standard deviations are small relative to means, indicating stable timing measurements. CGS shows slightly higher variance, possibly due to cache behavior.

**Performance ranking (for well-conditioned matrices)**

1. 🥇 **Householder QR** (`np.linalg.qr`): Fastest + most numerically stable
2. 🥈 **Classical GS**: Faster than MGS but less stable for ill-conditioned inputs
3. 🥉 **Modified GS**: Slower but more stable than CGS under near-dependency

**Actionable recommendations**

- **For production use**: Always prefer **`np.linalg.qr(A, mode='reduced')`** — it is 5–7× faster and uses numerically superior Householder reflections via optimized LAPACK.
- **For educational purposes**: CGS and MGS are useful for understanding orthogonalization but should not be used for performance-critical code.
- **For streaming/incremental orthogonalization**: MGS is preferred when you cannot access the full matrix at once, despite its runtime overhead.

````python
import numpy as np
import matplotlib.pyplot as plt

# Validate required data sources
if 'ort_diagnostics' not in globals():
    raise RuntimeError("`ort_diagnostics` not found. Re-run Computation 3 before plotting.")
if 'stress_results' not in globals():
    raise RuntimeError("`stress_results` not found. Re-run Computation 5 before plotting.")
if 'runtime_results' not in globals() or not runtime_results:
    raise RuntimeError("`runtime_results` not found. Re-run Computation 6 before plotting.")

# Helper to fetch diagnostics with graceful fallback
def get_diag_value(diag_dict, tags, key):
    for tag in tags:
        diag = diag_dict.get(tag)
        if diag and key in diag:
            return float(diag[key])
    return np.nan

# --- Orthogonality bar chart data ---
baseline_cgs = get_diag_value(ort_diagnostics, ['cgs', 'cgs_last', 'cgs_generated'], 'ort_err')
baseline_mgs = get_diag_value(ort_diagnostics, ['mgs', 'mgs_alt', 'mgs_generated', 'mgs_last'], 'ort_err')

# Choose the most ill-conditioned stress case (max cond)
stress_eps = max(stress_results.keys(), key=lambda eps: stress_results[eps]['cond'])
stress_entry = stress_results[stress_eps]
stressed_cgs = float(stress_entry['cgs']['ort_err'])
stressed_mgs = float(stress_entry['mgs']['ort_err'])

bar_labels = [
    'CGS baseline',
    'MGS baseline',
    f"CGS stressed (eps={stress_eps:.0e})",
    f"MGS stressed (eps={stress_eps:.0e})"
]
bar_values = [baseline_cgs, baseline_mgs, stressed_cgs, stressed_mgs]
bar_colors = ['#5e81ac', '#a3be8c', '#d08770', '#ebcb8b']

# --- Runtime plot data ---
methods = [
    ('cgs', 'Classical GS', '#5e81ac'),
    ('mgs', 'Modified GS', '#a3be8c'),
    ('householder', 'Householder QR', '#bf616a')
]
ks = np.array([entry['k'] for entry in runtime_results], dtype=float)
order = np.argsort(ks)
ks = ks[order]

fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Plot orthogonality bars (log scale)
ax0 = axes[0]
bars = ax0.bar(np.arange(len(bar_labels)), bar_values, color=bar_colors, width=0.65)
ax0.set_yscale('log')
ax0.set_ylabel(r'Frobenius orthogonality error $\|Q^T Q - I\|_F$')
ax0.set_title('Orthogonality vs. stress levels')
ax0.set_xticks(np.arange(len(bar_labels)))
ax0.set_xticklabels(bar_labels, rotation=20, ha='right')
for idx, bar in enumerate(bars):
    val = bar_values[idx]
    ax0.text(bar.get_x() + bar.get_width() / 2, val * 1.2, f"{val:.1e}", ha='center', va='bottom', fontsize=9)
ax0.grid(True, which='both', axis='y', linestyle='--', alpha=0.3)

# Plot runtime curves with std shading
ax1 = axes[1]
for method, label, color in methods:
    means = np.array([runtime_results[i]['methods'][method]['mean'] for i in order], dtype=float)
    stds = np.array([runtime_results[i]['methods'][method]['std'] for i in order], dtype=float)
    ax1.plot(ks, means, marker='o', color=color, label=label, linewidth=2)
    lower = np.clip(means - stds, a_min=0, a_max=None)
    upper = means + stds
    ax1.fill_between(ks, lower, upper, color=color, alpha=0.15)

ax1.set_xlabel('k (with n = 5k)')
ax1.set_ylabel('Mean runtime (seconds)')
ax1.set_xscale('log')
ax1.set_yscale('log')
ax1.set_title('Runtime scaling across methods')
ax1.grid(True, which='both', linestyle='--', alpha=0.3)
ax1.legend()

fig.suptitle('Section C · Visual Diagnostics', fontsize=16)
fig.tight_layout()
plt.show()
````

**Plot takeaways**

- Householder QR dominates runtime (5–7× faster) and preserves the same stability, reinforcing why `np.linalg.qr` is the production-ready choice.
- CGS remains slightly faster than MGS, but with no stability upside under stress; when streaming/incremental orthogonalization is needed, the runtime penalty of MGS is the cost of robust behavior.
